```python
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Load the data
df = pd.read_csv('processed_employee_attrition_data.csv')

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("Employee Attrition Dashboard"),
    
    # Dropdown for selecting visualization
    html.Label("Select Visualization:"),
    dcc.Dropdown(
        id='visualization-dropdown',
        options=[
            {'label': 'Attrition Count', 'value': 'attrition_count'},
            {'label': 'Monthly Income vs Attrition', 'value': 'monthly_income_vs_attrition'},
            {'label': 'Age Distribution', 'value': 'age_distribution'},
            {'label': 'Correlation Matrix', 'value': 'correlation_matrix'},
            {'label': 'Attrition by Department', 'value': 'attrition_by_department'},
            {'label': 'Attrition by Gender', 'value': 'attrition_by_gender'},
            {'label': 'Attrition by Job Role', 'value': 'attrition_by_job_role'},
            {'label': 'Attrition by Education Field', 'value': 'attrition_by_education_field'},
            {'label': 'Attrition by Marital Status', 'value': 'attrition_by_marital_status'},
            {'label': 'Monthly Income Distribution', 'value': 'monthly_income_distribution'}
        ],
        value='attrition_count'
    ),
    
    # Graph to display the visualization
    dcc.Graph(id='visualization-graph')
])

# Callback to update the graph based on the selected visualization
@app.callback(
    Output('visualization-graph', 'figure'),
    Input('visualization-dropdown', 'value')
)
def update_graph(selected_visualization):
    if selected_visualization == 'attrition_count':
        fig = px.histogram(df, x='Attrition', title='Attrition Count')
    elif selected_visualization == 'monthly_income_vs_attrition':
        fig = px.box(df, x='Attrition', y='MonthlyIncome', title='Monthly Income vs Attrition')
    elif selected_visualization == 'age_distribution':
        fig = px.histogram(df, x='Age', nbins=30, title='Age Distribution')
    elif selected_visualization == 'correlation_matrix':
        corr_matrix = df.corr()
        fig = px.imshow(corr_matrix, title='Correlation Matrix')
    elif selected_visualization == 'attrition_by_department':
        fig = px.histogram(df, x='Department_Research & Development', color='Attrition', barmode='group', title='Attrition by Department')
    elif selected_visualization == 'attrition_by_gender':
        fig = px.histogram(df, x='Gender', color='Attrition', barmode='group', title='Attrition by Gender')
    elif selected_visualization == 'attrition_by_job_role':
        fig = px.histogram(df, y='JobRole_Sales Executive', color='Attrition', barmode='group', title='Attrition by Job Role')
    elif selected_visualization == 'attrition_by_education_field':
        fig = px.histogram(df, x='EducationField_Life Sciences', color='Attrition', barmode='group', title='Attrition by Education Field')
    elif selected_visualization == 'attrition_by_marital_status':
        fig = px.histogram(df, x='MaritalStatus_Single', color='Attrition', barmode='group', title='Attrition by Marital Status')
    elif selected_visualization == 'monthly_income_distribution':
        fig = px.histogram(df, x='MonthlyIncome', nbins=30, title='Monthly Income Distribution')
    else:
        fig = px.histogram(df, x='Attrition', title='Attrition Count')
    
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
